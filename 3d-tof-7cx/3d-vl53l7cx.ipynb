```python
import numpy as np  
import matplotlib.pyplot as plt  
from matplotlib.animation import FuncAnimation  
from mpl_toolkits.mplot3d import Axes3D  

# Settings  
resolution = 8  # Assuming an 8x8 grid for ToF sensor  
x = np.arange(resolution)  
y = np.arange(resolution)  
X, Y = np.meshgrid(x, y)  

# Generate initial random data  
data = np.random.randint(10, 1000, size=(resolution, resolution))  

# Figure setup with two subplots: 3D Plot and Heatmap  
fig = plt.figure(figsize=(12, 6))  

# 3D Plot  
ax1 = fig.add_subplot(121, projection='3d')  
surface = ax1.plot_surface(X, Y, data, cmap='viridis')  

# Heatmap  
ax2 = fig.add_subplot(122)  
heatmap = ax2.imshow(data, cmap='viridis', interpolation='nearest')  
colorbar = plt.colorbar(heatmap, ax=ax2)  
ax2.set_title('Heatmap')  

# Update function for real-time plotting  
def update(frame):  
    global data  
    # Simulate new data (replace this with real sensor data in practice)  
    data = np.random.randint(10, 1000, size=(resolution, resolution))  
    
    # Update the 3D surface  
    ax1.clear()  
    surface = ax1.plot_surface(X, Y, data, cmap='viridis')  
    ax1.set_title('Real-Time 3D Plot')  
    ax1.set_xlabel('X')  
    ax1.set_ylabel('Y')  
    ax1.set_zlabel('Z (Distance)')  

    # Update the Heatmap  
    heatmap.set_data(data)  
    ax2.set_title('Heatmap')  
    ax2.set_xlabel('X')  
    ax2.set_ylabel('Y')  
    colorbar.update_normal(heatmap)  

# Animate  
ani = FuncAnimation(fig, update, interval=500)  # Update every 500 ms  

# Display the plots  
plt.tight_layout()  
plt.show()  
```
